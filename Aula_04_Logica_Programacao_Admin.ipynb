# **Exercício 1:** Lógica e Condicionais (O Semáforo das Vendas)


## **Exemplo 01:**
Um administrador precisa de um programa que classifique o desempenho diário de uma loja de conveniência.

Tarefa: Peça para o usuário digitar o valor total de vendas do dia.

Se o valor for menor que R$ 1.000, imprima: "Desempenho Baixo: Revisar estoque e promoções."

Se estiver entre R$ 1.000 e R$ 3.000, imprima: "Desempenho Médio: Dentro da normalidade."

Se for maior que R$ 3.000, imprima: "Desempenho Alto: Parabéns à equipe!"

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Resposta 1: Condicionais ---
vendas_dia = float(input("Digite o valor das vendas: "))
if vendas_dia < 1000:
    print("Desempenho Baixo: Revisar estoque e promoções.")
elif vendas_dia <= 3000:
    print("Desempenho Médio: Dentro da normalidade.")
else:
    print("Desempenho Alto: Parabéns à equipe!")
```

# **Exercício 2:** Comandos Iterativos (Calculando a Média de Itens)


## **Exemplo 02:** Você tem uma lista com a quantidade de produtos vendidos em uma semana: vendas = [45, 60, 32, 85, 90, 120, 75].

Tarefa: Use um loop for para percorrer a lista e somar todas as unidades. Ao final, calcule e exiba a média de vendas diárias.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# --- Resposta 2: Loop for ---
vendas = [45, 60, 32, 85, 90, 120, 75]
soma_vendas = 0
for v in vendas:
    soma_vendas = soma_vendas + v
media = soma_vendas / len(vendas)
print(f"Total vendido: {soma_vendas} | Média diária: {media:.2f}")
```

# **Exercício 3:** Pandas e Arquivos (Leitura e Estrutura)


## **Exemplo 03:** Suponha que você recebeu um arquivo chamado relatorio_vendas.xlsx com colunas: Produto, Quantidade e Preço Unitário.

# **Exercício 5:** Visualização com Seaborn (Gráfico de Linhas e Barras)
Com os dados de receita prontos:



## **Exemplo 05:** Tarefa A: Crie um gráfico de barras usando sns.barplot para mostrar a Receita Total por Produto.

Tarefa B: Imagine que você tem uma coluna Data. Crie um gráfico de linhas com sns.lineplot para visualizar a evolução das vendas ao longo do tempo. Adicione títulos e nomes nos eixos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Exercício 5: Geração de Gráficos (Versão Atualizada) ---

# Configurando o estilo visual
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 6))

# 1. Gráfico de Barras: Receita por Produto
plt.subplot(1, 2, 1)

# Atribuímos 'Produto' ao parâmetro 'hue'
# Definimos 'legend=False' para não poluir o gráfico com uma legenda redundante
sns.barplot(
    data=df_vendas,
    x='Produto',
    y='Receita Total',
    estimator=sum,
    errorbar=None,
    hue='Produto',
    legend=False,
    palette='magma'
)

plt.title('Receita Total por Produto')
plt.xticks(rotation=45)

# 2. Gráfico de Linhas: Evolução Temporal
plt.subplot(1, 2, 2)
if 'Data' in df_vendas.columns:
    df_vendas['Data'] = pd.to_datetime(df_vendas['Data'])
    sns.lineplot(data=df_vendas, x='Data', y='Receita Total', marker='o', color='teal')
    plt.title('Evolução das Vendas no Tempo')
    plt.xticks(rotation=45)
else:
    print("Coluna 'Data' não encontrada para o gráfico de linhas.")

plt.tight_layout()
plt.show()
```
